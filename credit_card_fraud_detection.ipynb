Credit Card Fraud Detection Using Logistic Regression

```python
import numpy as np 
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
credit_card_data = pd.read_csv('creditcard.csv')
```

```python
credit_card_data.head()
```

  0 -> Normal Transaction
  
  1 -> Fraud Transaction

```python
credit_card_data.isnull().sum()
```

```python
credit_card_data['Class'].value_counts()
```

```python
# the data is unbalanced to we balance it first 
```

```python
# seprating data for analysis

normal_trans_data = credit_card_data[credit_card_data.Class == 0]
fraud_data = credit_card_data[credit_card_data.Class == 1]
```

```python
normal_trans_data.Amount.describe()
```

```python
fraud_data.Amount.describe()
```

```python
# cpmaring the values of both transcation
credit_card_data.groupby('Class').mean()
```

```python
# there is wide difference btw mean of both value so we need to balance the data to  make this difference as less as possible 
```

```python
# Using UnderSampling Technique to solve this problem
# for this we are building a sample dataset containing similiar distribution of normal transaction and fraud transcation 
```

```python
fraud_data.shape
```

```python
normal_data_sample = normal_trans_data.sample(n=492)
```

```python
# concatetinating two data set 
```

```python
new_dataset = pd.concat([normal_data_sample,fraud_data], axis = 0) #axixs = 0 means the data will me added row wise if it will be 1 it will add column wise
```

```python
new_dataset.head()
```

```python
new_dataset.shape
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
## Spliting the data into feature and target
```

```python
X = new_dataset.drop(columns='Class',axis =1)
Y = new_dataset['Class']
```

```python
# spliting data into training and testing data 
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size= 0.2, stratify=Y, random_state = 40)
```

```python
#Model training
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
# checking accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction,Y_train)
print('Accuracy Score',training_data_accuracy*100)
```

```python
# checking accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction,Y_test)
print('Accuracy Score',test_data_accuracy*100)
```
